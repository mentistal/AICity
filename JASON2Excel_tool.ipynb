```python
import os
import cv2
import numpy as np
import json
import pandas as pd
```

```python
# Function to load annotations from a JSON file
def load_annotations(annotation_file):
    with open(annotation_file, 'r') as f:
        return json.load(f)
```

```python
annotation_files = ['D:\Projects\AICity\Tiles\Tiles\instances_1_2_3.json',
                    'D:\Projects\AICity\Tiles\Tiles\instances_4_5_6.json',
                    'D:\Projects\AICity\Tiles\Tiles\instances_7_8_9.json',
                    'D:\Projects\AICity\Tiles\Tiles\instances_10_11_12.json']

output_dir = 'D:\Projects\AICity\datasheet'

# Ensure the output directory exists
os.makedirs(output_dir, exist_ok=True)
```

```python
DataSheet = pd.DataFrame()
for i,file in enumerate(annotation_files):
    annotation_data=load_annotations(file)
    for j,annotation in enumerate(annotation_data['annotations']):
        id = annotation['id']
        image_idx = int(annotation['image_id'])+3*i
        region = str(annotation['segmentation'])
        area = str(annotation['area'])
        bbox = str(annotation['bbox'])
        a_colour = annotation['attributes']['Colours']
        a_texture = annotation['attributes']['Textures']
        a_score = str(annotation['attributes']['Score'])
        a_materials = annotation['attributes']['Materials']
        a_occluded = annotation['attributes']['occluded']
        new_data = pd.DataFrame([{'ID':f'{image_idx}-{id}',
                    'Area':area, 
                    'BBox':bbox, 
                    'Colour':a_colour, 
                    'Texture':a_texture, 
                    'Meterial':a_materials, 
                    'Score':a_score, 
                    'Region':region}])
        DataSheet=pd.concat([DataSheet,new_data], ignore_index=True)
       
        
   
   
DataSheet
```

```python
with pd.ExcelWriter('AICity_data_sheet.xlsx') as writer:
    DataSheet.to_excel(writer, sheet_name='AICity', index=False)
```
